```python
import numpy as np
import scipy.io
from utils import *
import matplotlib.pyplot as plt
import pandas as pd
```

### Load RNP SSI Results

```python
base_path = 'd:/data/rnp/'
ob_skaggs = load_mat_file(Path(base_path) / 'ob_skaggs_dist.mat') 
hc_skaggs = load_mat_file(Path(base_path) / 'hc_skaggs_dist.mat') 
```

```python
col_names = ['mouse_id', 'session_id', 'cluster_id', 'si', 'ssi', 'p_val']
ob_skaggs_df = pd.DataFrame(ob_skaggs, columns=col_names)
hc_skaggs_df = pd.DataFrame(hc_skaggs, columns=col_names)
```

### Filter Clusters with SSI

```python
# Get rows with significant SSI (p_val < 0.05)
significant_ob = ob_skaggs_df[(ob_skaggs_df['ssi'] >= 3) & (ob_skaggs_df['si'] >= 0.5)]
significant_hc = hc_skaggs_df[(hc_skaggs_df['ssi'] >= 3) & (hc_skaggs_df['si'] >= 0.5)]
```

```python
significant_ob.to_csv(Path(base_path, 'significant_ob.csv'), index=False)
significant_ob
```
