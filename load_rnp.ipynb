```python
import numpy as np
import scipy.io
from utils import *
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def load_mat_file(filepath, variable_name=None):
    # Load the .mat file
    mat_data = scipy.io.loadmat(filepath)
    # Remove metadata keys that MATLAB adds
    mat_data = {key: value for key, value in mat_data.items() 
                if not key.startswith('__')}
    mat_data = mat_data[list(mat_data.keys())[0]]
    # If a specific variable is requested, return only that variable 
    return mat_data
```

```python
base_path = 'd:/data/rnp/'
mtrack_paths = get_file_paths(base_path, keyword='mtrack', extension='mat', print_paths=True)
```

```python
# Load a file
mtrack = load_mat_file(mtrack_paths[0]) 

# Convert to Pandas and name columns
columns = {0: 'frame', 1: 'head_x', 2: 'body_x', 3: 'head_y', 4: 'body_y'}
mtrack = pd.DataFrame(mtrack)
mtrack = mtrack.rename(columns = columns)
```
